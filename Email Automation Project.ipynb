```python
#Adding GUI    

#Initializing the program
root = Tk()
root.title('Email Automation')
root.geometry('1440x900')

#Loading images
confirm_img = PhotoImage(file = "images/confirm_email.png")
send_img = PhotoImage(file = "images/send_email.png")


#To open new window when Get Started button is clicked:
def openNewWindow():
     
    # Toplevel object which will be treated as a new window
    newWindow = Toplevel(root)
 
    # sets the title of the Toplevel widget
    newWindow.title("Email Automation Bot")
 
    # sets the geometry of toplevel
    newWindow.geometry("700x600")
    
    #Sets background color
    newWindow.configure(bg='#EFF6F0')
 
    # A Label widget to show in toplevel
    Label(newWindow, text ="Enter your Email address:",font="Calibri", bg="#EFF6F0").pack()
    
    #Label Creation
    lbl = Label(newWindow, text = "")
    lbl.pack()
    

    def printInput():
        inp = inputtxt.get(1.0, "end-1c")
        get_email_info(inp)
        show_details()

    def show_details():
        Label(newWindow, text =f'To: {list[0]}', font = ("Calibri", 10),bg= "#EFF6F0").pack()
        Label(newWindow, text =f'Subject: {list[1]}', font = ("Calibri", 10)).pack()
        Label(newWindow, text =f'Content: {list[2]}', font = ("Calibri", 10)).pack()
      
        
  
    # TextBox to get email address
    inputtxt = Text(newWindow,height = 1.5,width = 25, font = ("Calibri", 12), borderwidth=0)

    inputtxt.pack()

    # Confirm Button Creation
    confirm_label= Label(image=confirm_img)
    button = Button(newWindow, image=confirm_img, borderwidth=0, bg="#EFF6F0", command = printInput)
    button.pack(pady=30)
    
    #Function to send email
    def final_send():
        send_email_here(list)
        
    #Send Button Creation    
    send_label = Label(image=send_img)
    send_button = Button(newWindow,image=send_img,command = final_send, borderwidth=0,bg="#EFF6F0")
    send_button.pack(pady=70)
    

#Main window background image
image = Image.open('images/Frame 334.png')
image = ImageTk.PhotoImage(image)
image_label = Label(image = image)
image_label.image = image
image_label.grid(column=1, row=0)

# Get Started Button Creation
start_btn = ImageTk.PhotoImage(file = "images/Get Started.png")
button = Button(
    root,
    image=start_btn,
    borderwidth=0,
    bg="#EFF6F0",
    command=openNewWindow,
)
button.place(x=107, y=625)

root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from PIL import Image, ImageTk
import smtplib
import speech_recognition as sr
import pyttsx3
from email.message import EmailMessage
import time

listener = sr.Recognizer()
engine = pyttsx3.init()

#To store email details
list = []


def talk(text):
    engine.say(text)
    engine.runAndWait()


def get_info():
    try:
        with sr.Microphone() as source:
#             print('listening...')
            voice = listener.listen(source)
            info = listener.recognize_google(voice)
            return info.lower()
    except:
        pass


def send_email(receiver, subject, message):
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    # Make sure to give app access in your Google account
    server.login('SENDER_EMAIL_ADDRESS', 'PASSWORD')
    email = EmailMessage()
    email['From'] = 'SENDER_EMAIL_ADDRESS'
    email['To'] = receiver
    email['Subject'] = subject
    email.set_content(message)
    server.send_message(email)



def get_email_info(email_to):
    receiver = email_to
    list.append(receiver)
    talk('What is the subject of your email?')
    subject = get_info()
    list.append(subject)
    talk('Tell me the text in your email')
    message = get_info()
    talk('Please click on send button to send the email after checking the content of the generated email')
    list.append(message)

#list has all the details entered by the user    
def send_email_here(list):
    send_email(list[0], list[1], list[2])
    talk('Your email is sent')
    talk('Thankyou for using this bot')
    

#Adding GUI    

#Initializing the Root window
root = Tk()
root.title('Email Automation')
root.geometry('1440x900')

#Loading images
confirm_img = PhotoImage(file = "confirm_email.png")
send_img = PhotoImage(file = "send_email.png")


#To open new window when Get Started button is clicked:
def openNewWindow():
     
    # Toplevel object which will be treated as a new window
    newWindow = Toplevel(root)
 
    # sets the title of the Toplevel widget
    newWindow.title("Email Automation Bot")
 
    # sets the geometry of toplevel
    newWindow.geometry("700x600")
    
    #Sets background color
    newWindow.configure(bg='#EFF6F0')
 
    # A Label widget to show in toplevel
    Label(newWindow, text ="Enter your Email address:",font="Calibri", bg="#EFF6F0").pack()
    
    #Label Creation
    lbl = Label(newWindow, text = "")
    lbl.pack()
    

    def printInput():
        inp = inputtxt.get(1.0, "end-1c")
        get_email_info(inp)
        show_details()

    def show_details():
        Label(newWindow, text =f'To: {list[0]}', font = ("Calibri", 12),bg= "#EFF6F0").pack()
        Label(newWindow, text =f'Subject: {list[1]}', font = ("Calibri", 12),bg= "#EFF6F0").pack()
        Label(newWindow, text =f'Content: {list[2]}', font = ("Calibri", 12),bg= "#EFF6F0").pack()
      
        
  
    # TextBox to get email address
    inputtxt = Text(newWindow,height = 1.5,width = 25, font = ("Calibri", 12), borderwidth=0)

    inputtxt.pack()

    # Confirm Button Creation
    confirm_label= Label(image=confirm_img)
    button = Button(newWindow, image=confirm_img, borderwidth=0, bg="#EFF6F0", command = printInput)
    button.pack(pady=35)
    
    #Function to send email
    def final_send():
        send_email_here(list)
        
    #Send Button Creation    
    send_label = Label(image=send_img)
    #time.sleep() is used to close subsequent windows with a time gap of 1.5 seconds.
    send_button = Button(newWindow,image=send_img,command = lambda: [final_send(), newWindow.destroy(), time.sleep(1.5), root.destroy()], borderwidth=0,bg="#EFF6F0")
    send_button.place(x=265, y=325)
    

    
#Main window background image
image = Image.open('Frame 334.png')
image = ImageTk.PhotoImage(image)
image_label = Label(image = image)
image_label.image = image
image_label.grid(column=1, row=0)

# Get Started Button Creation
start_btn = ImageTk.PhotoImage(file = "Get Started.png")
button = Button(
    root,
    image=start_btn,
    borderwidth=0,
    bg="#EFF6F0",
    command=openNewWindow,
)
button.place(x=107, y=625)

root.mainloop()
```
